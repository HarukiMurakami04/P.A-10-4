```python
num = int(input(f"Digite um numero"))

if num % 2 == 0:

  print(f"{num} é par")
else:
  print(f"{num} é impar")
```

```python
num = int(input(f"Digite a nota"))

if num >= 7:
  print(f"voce esta aprovado")
elif num >= 5:
  print(f"por pouco, voce passou")
else:
  print(f"voce reprovou")
```

```python
num = float(input(f"Digite o preco"))

if num >= 500:
  print(f"voce tem um desconto de 20%")
  print(f"ou seja voce voce pagara {num*0.8}")
elif num >= 101:
  print(f"voce tera um desconto de 10%")
  print(f"ou seja, voce pagara {num*0.9}")
else:
  print(f"voce nao tera desconto")
  print(f"ou seja voce pagara {num}")
```
