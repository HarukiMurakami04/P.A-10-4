```python
num = int(input(f"digite um numero:"))

if num % 2 == 0:

  print(f"{num} é par")
else:
  print(f"{num} é impar")
```

```python
bim1 = int(input(f"digite a nota do primeiro bimestre:"))
bim2 = int(input(f"digite a nota do segundo bimestre:"))
bim3 = int(input(f"digite a nota do terceiro bimestre:"))
bim4 = int(input(f"digite a nota do quarto bimestre:"))
num = (bim1+bim2+bim3+bim4)/4

if num >= 7:
  print(f"voce esta aprovado")
  print(f"voce tirou {num}")
elif num >= 5:
  print(f"por pouco, voce passou")
  print(f"voce tirou {num}")
else:
  print(f"voce reprovou")
  print(f"voce tirou {num}")
```

```python
num = float(input(f"Digite o preco"))

if num >= 500:
  print(f"voce tem um desconto de 20%")
  print(f"ou seja voce voce pagara {num*0.8}")
elif num >= 101:
  print(f"voce tera um desconto de 10%")
  print(f"ou seja, voce pagara {num*0.9}")
else:
  print(f"voce nao tera desconto")
  print(f"ou seja voce pagara {num}")
```
